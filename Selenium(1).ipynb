```python
from selenium import webdriver

browser = webdriver.Chrome() # 괄호 안은 exe파일의 경로를 써 주는 부분
browser.get("http://www.naver.com")
```

```python
elem = browser.find_element_by_class_name("link_login")
```

```python
# elem.click() : 클릭!
# browser.back() : 뒤로가기
# browser.forward(): 앞으로 가기
# browser.refresh(): 새로고침
```

```python
elem = browser.find_element_by_id("query")
```

```python
from selenium.webdriver.common.keys import Keys
# 글자를 입력하기 위한(엔터를 위한) import

#elem.send_keys("테스트")
#elem.send_keys(Keys.ENTER)

# 특정 태그를 찾는법
#elem = browser.find_element_find_by_tag_name("a") # 한 개
elem = browser.find_elements_by_tag_name("a") # 모든 a 태그

for e in elem:
    print(e.get_attribute("href")) #속성찾기, 뷰티풀 수프에서는 대괄호
```
