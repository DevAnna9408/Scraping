```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import requests
from bs4 import BeautifulSoup
browser = webdriver.Chrome()
browser.get("https://play.google.com/store/movies?hl=ko&gl=US")
browser.maximize_window()
```

```python
# 모니터의 세로 해상도만큼 스크롤 내리기
#browser.execute_script("window.scrollTo(0, 1080)")
# 페이지의 가장 아래로 스크롤 내리기
browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

interval = 3

#현재 문서의 높이를 가져와서 저장
prev_height = browser.execute_script("return document.body.scrollHeight")

# 반복
while True:
    # 스크롤을 가장 아래로 내린 다음
    browser.execute_script("return document.body.scrollHeight")
    
    # 페이지 로딩 대기
    time.sleep(interval)
    
    # 현재 문서 높이를 가져와서 저장
    curr_height = browser.execute_script("return document.body.scrollHeight")
    if curr_height == prev_height:
        break
        
    prev_height = curr_height
    
print('스크롤 완료')

soup = BeautifulSoup(browser.page_source, "lxml")
# 구글무비 접속 한 후 페이지를 다 내린 HTML정보를 가져오는 겁니다.

movies = soup.find_all("div", attrs = {"class":"WHE7ib mpg5gc"})

url = "https://play.google.com/store/movies?hl=ko&gl=US"

# headers = {
#     "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36",
#     "Accept-Language":"ko-KR,ko"
#     }

res = requests.get(url)
res.raise_for_status()

for movie in movies:
    # 영화 제목
    title = movie.find("div", attrs = {"class":"WsMG1c nnK0zc"}).get_text()
    
    # 할인 되지 않은 가격
    original_price = movie.find("span", attrs = {"class":"VfPpfd ZdBevf i5DZme"}).get_text()
        
    # 할인 된 가격
    real_price = movie.find("span", attrs = {"class":"SUZt4c djCuy"})
    
    # 영화 링크, https://play.google.com + link
    link = movie.find("a", attrs = {"class":"JC71ub"})["href"]
    
    print(f"제목 : {title}")
    print(f"할인 전 금액 : {original_price}")
    print(f"할인 후 금액 : {real_price}")
    print("링크 : ", "https://play.google.com" + link)
    print('-'*120)
    
browser.quit()
```
